# Oszcillátorok numerikus vizsgálata

$$ \def\dd{{\textrm{d}}} $$

Oszcillátoroknak olyan dinamikai rendszereket nevezünk, melyek mozgása az idő periodikus vagy kvázi-periodikus függvényeként írható le. Az egydimenziós oszcillátor mozgásegyenlete

$$ m \frac{\dd^2x}{\dd t^2} = - \frac{\dd V(x)}{\dd x} $$

alakban írható. Egy szabad oszcillátor teljes energiája mindaddig megmarad, amíg a rendszerben nem hat súrlódási erő (csillapított oszcillátor), illetve nem hatunk rá külső erővel (gerjesztett oszcillátor). Az oszcillátor teljes energiája a kinetikus és a potenciális energia összege, azaz

$$ E_{\textrm{tot}} = \frac{m v^2}{2} + V(x)$$

## Harmonikus oszcillátor

Amennyiben a $V(x)$ potenciál négyzetes, abban az esetben harmonikus oszcillátorról beszélünk:

$$\begin{eqnarray}
    V(x) & = & \frac{1}{2}kx^2 \\
    m \frac{\dd^2x}{\dd t^2} & = & - k x.
\end{eqnarray}$$

A harmonikus oszcillátor mozgásegyenlete analitikusan is megoldható. A megoldás

$$ x(t) = A \cos( \omega t + \phi ) $$

alakú, ahol $A$ az amplitúdó - vagyis a legnagyobb kitérés -, $\omega$ a körfrekvencia és $\phi$ a kezdőfázis $t = 0$-ban. A harmonikus oszcillátor körfrekvenciáját a potenciálban szereplő $k$ állandó és az $m$ tömeg jellgű mennyiség határozza meg:

$$ \omega = \sqrt{\frac{k}{m}} = \frac{2\pi}{T}, $$

ahol $T$ a pediódusidő, melyből $f = 1/T$ módon kaphatjuk meg a frekvenciát.

## Anharmonikus oszcillátor

Az anharmonikus abban különböznek a harmonikus oszcillátortól, hogy a mozgásukat meghatározó potenciál nem négyzetes alakú. Néhány lehetséges anharmonikus potenciál:

* Negyedrendű potenciál

$$ V(x) = \frac{1}{4}kx^4 $$

* Lennard-Jones potenciál

$$ V(x) = V_0 \left[ \left( \frac{x_0}{x} \right)^{12} - 2 \left( \frac{x_0}{x} \right)^{6} \right] $$

## Csillapított és gerjesztett oszcillátor

A csillapított és/vagy gerjesztett oszcillátorok mozgásegyenletében megjelenik egy sebességtől függő és egy további, többnyire időfüggő külső erőtag is:

$$ m \frac{\dd^2x}{\dd t^2} = - \frac{\dd V(x)}{\dd x} - c \frac{\dd x}{\dd t} + F(t) $$

Csillapított és szinuszosan gerjesztett oszcillátorok viselkedése analitikusan is egyszerűen vizsgálható. Az alábbi feladatokban ezért ezektől eltérő rendszereket vizsgálunk numerikus módszerek segítségével.

## 1. feladat

Számítsuk ki a fent felírt két anharmonikus potenciálhoz tartozó erőt.

Negyedrendű anharmonikus potenciálhoz tartozó erő:

$$ F(x) = -kx^4 $$

Lennard-Jones potenciálhoz tartozó erő:

$$ F(x) = -12 V_0 \left[ \left( \frac{x_0^{12}}{x^{13}} \right) - \left( \frac{x_0^{6}}{x^{7}} \right) \right] $$


## 2. feladat

Írjunk olyan függvényeket, melyek kiszámítják a harmonikus potenciált, az ehhez tartozó erőt, illetve az 1. feladatban felírt anharmonikus potenciálokat és erőket. Ábrázoljuk a potenciálokat és az erőket grafikonon!

```python
# potenciálokat és erőket kiszámító függvények :

def v_harm(x,k):
    V = (1/2)*k*x**2
    return V

def F_harm(x,k):
    F = -k*x
    return F

def v_4rend(x,k):
    V = (1/4)*k*x**4
    return V

def F_4rend(x,k):
    F = -k*x**3
    return F

def v_lj(x,v0,x0):
    V = v0*(x0/x)**12 - 2*(x0/x)**6
    return V

def F_lj(x,v0,x0):
    F = -12*v0*((x0**12)/(x**13)-(x0**6)/(x**7))
    return F
```

```python
# potenciálok és erők ábrázolása :

import matplotlib.pyplot as plt
import numpy as np

plt.title('Harmonikus oszcillátor potenciálja')
x = np.linspace(-10,10,100)
plt.plot(x,v_harm(x,10),label='Potenciál')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
```

```python
plt.title('Harmonikus oszcillátor potenciáljához tartozó erő')
plt.plot(x,F_harm(x,10),label='Erő')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
```

```python
plt.title('Negyedrendű anharmonikus oszxillátor potenciálja')
plt.plot(x,v_4rend(x,10),label='Potenciál')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
```

```python
plt.title('Negyedrendű anharmonikus oszcillátor potenciáljához tartozó erő')
plt.plot(x,F_4rend(x,10),label='Erő')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
```

```python
plt.title('Lennard-Jones potenciál')
a = np.linspace(1.3,3,1000)
plt.plot(a,v_lj(a,1,1.5),label='Potenciál')
plt.xlabel('x/$x_0$')
plt.ylabel('V/$V_0$')
plt.legend()
plt.grid()
```

```python
plt.title('Lennard-Jones potenciálhoz tartozó erő')
a = np.linspace(1,3,1000)
plt.plot(a,F_lj(a,1,1.5),label='Erő')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
```

## 3. feladat

Integráljuk a harmonikus és Lennard-Jones-potenciállal meghatározott oszcillátorok mozgásegyenletét egyszerű negyed rendű Runge-Kutta-módszerrel vagy a Dormand-Prince-féle 5(4)-ed rendű, adaptív lépéshossz-választásos módszerrel. Az integráláshoz használjunk szabadon elérhető programcsomagot, pl.:

* python: scipy.integrate.RK45
* octave: ode45
* C: GSL ODE csomag
* C++: Boost.Numeric.Odeint csomag

Ábrázoljuk az egyenlet változóinak időfejlődését, a teljes energiát és a lépéshosszt az idő függvényében!

```python
# harmonikus oszcillátor mozgásegyenletének integrálása :
from scipy.integrate import RK45

def harmosc(t,y):
    k = 10
    m = 20
    x = y[0]
    v = y[1]
    dx = v
    dv = (1/m)*F_harm(x,k)
    return dx, dv

t0=0
y0=(5,0)
tmax = 75
t = []
y = []

rk = RK45(harmosc,t0,y0,tmax, rtol=1e-3)
t.append(rk.t)
y.append(rk.y)

while(rk.t < tmax):
    rk.step()
    t.append(rk.t)
    y.append(rk.y)
    
y = np.array(y)
t = np.array(t)
```

```python
# harmonikus oszcillátor ábrázolása :
plt.plot(t,y[:,0],'+--')
plt.xlabel('t (idő)')
plt.ylabel('y (kitérés)')
plt.title('Harmonikus oszcillátor')
plt.grid()
```

```python
# lennard-jones potenciál mozgásegyenletének integrálása :
def lj(t,y):
    m = 200
    x = y[0]
    v = y[1]
    x0 = 1.9
    v0 = 0.001
    dx = v
    dv = (1/m)*F_lj(x,v0,x0)
    return dx, dv

tj = []
yj = []
y0=[2,0]
ttmax = 500

rkj = RK45(lj,t0,y0,tmax,rtol=1e-3)
tj.append(rkj.t)
yj.append(rkj.y)

while(rkj.t < ttmax):
    rkj.step()
    tj.append(rkj.t)
    yj.append(rkj.y)
    
yj = np.array(yj)
tj = np.array(tj)
```

```python
# lennard-jones potenciál ábrázolása :
plt.plot(tj,yj[:,0],'+--')
plt.xlabel('t (idő)')
plt.ylabel('y (kitérés)')
plt.title('Lennard-Jones potenciál')
plt.grid()
# ez az ábra még mindig nem jó
```

```python
# negyedrendű anharmonikus oszcillátor :
def negy(t,y):
    k = 10
    m = 1
    x = y[0]
    v = y[1]
    dx = v
    dv = (1/m)*F_4rend(x,k)
    return dx, dv

t0=0
y0=(5,0)
tmax_uj = 200
t_uj = []
y_uj = []
rk_uj = RK45(negy,t0,y0,tmax_uj,rtol=1e-2)
t_uj.append(rk_uj.t)
y_uj.append(rk_uj.y)
while(rk_uj.t < tmax_uj):
    rk_uj.step()
    t_uj.append(rk_uj.t)
    y_uj.append(rk_uj.y)
    
y_uj = np.array(y_uj)
t_uj = np.array(t_uj)

plt.plot(t_uj,y_uj[:,0])
plt.xlabel('t (idő)')
plt.ylabel('y (kitérés)')
plt.title('Negyedrendű anharmonikus oszcilátor')
plt.grid()
# ezen inkább az látszik, hogy előbb-utóbb lecseng, de egy lejjebbi ábrán az anharmonikusság is szépen kivehető
```

## 4. feladat

Interpoláljuk a 3. feladatból kapott megoldásokat időben egyenletes lépésközzel, lineárisan és köbös spline-ok segítségével!

```python
# harmonikus oszcillátor, lineáris spline-okkal :
from scipy.interpolate import interp1d
ip = interp1d(t,y[:,0],kind='linear')
tt = np.linspace(0,tmax,1000)
plt.plot(tt,ip(tt))
plt.xlabel('t (idő)')
plt.ylabel('y (kitérés)')
plt.title('Harmonikus oszcillátor - lineáris spline-okkal')
plt.grid()
```

```python
# harmonikus oszcillátor, köbös spline-okkal :
ips = interp1d(t,y[:,0],kind='cubic')
tts = np.linspace(0,tmax,1000)
plt.plot(tts,ips(tts))
plt.xlabel('t (idő)')
plt.ylabel('y (kitérés)')
plt.title('Harmonikus oszcillátor - köbös spline-okkal')
plt.grid()
```

```python
# lennard-jones, lineáris spline-okkal :
ip = interp1d(tj,yj[:,0],kind='linear')
tt = np.linspace(0,tmax,100)
plt.plot(tt,ip(tt))
plt.xlabel('t (idő)')
plt.ylabel('y (kitérés)')
plt.title('Lennard-Jones - lineáris spline-okkal')
plt.grid()
```

```python
# lennard-jones potenciál, köbös spline-okkal :
iplj = interp1d(tj,yj[:,0],kind='cubic')
ttlj = np.linspace(0,tmax,100)
plt.plot(ttlj,iplj(ttlj))
plt.xlabel('t (idő)')
plt.ylabel('y (kitérés)')
plt.title('Lennard-Jones - köbös spline-okkal')
plt.grid()
```

```python
# Negyedrendű anharmonikus oszcillátor lineáris spline-okkal :
ipnl = interp1d(t_uj,y_uj[:,0],kind='linear')
tnl = np.linspace(0,tmax_uj,200)
plt.plot(tnl,ipnl(tnl))
plt.xlabel('t (idő)')
plt.ylabel('y (kitérés)')
plt.title('Negyedrendű anharmonikus oszcillátor - lineáris spline-okkal')
plt.grid()
# nagyobb lépésközt állítottam be, hogy itt látszódjon, hogy anharmonikus
```

```python
# Negyedrendű anharmonikus oszcillátor köbös spline-okkal :
ipnk = interp1d(t_uj,y_uj[:,0],kind='cubic')
tnk = np.linspace(0,tmax_uj,200)
plt.plot(tnk,ipnk(tnk))
plt.xlabel('t (idő)')
plt.ylabel('y (kitérés)')
plt.title('Negyedrendű anharmonikus oszcillátor - köbös spline-okkal')
plt.grid()
```
